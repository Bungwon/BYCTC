```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import joblib
```

```python
train_df = pd.read_csv("train.csv")

train_df
```

```python
label_encoder_sex = LabelEncoder()
train_df['Sex_encoded'] = label_encoder_sex.fit_transform(train_df['Sex'])

label_encoder_embark = LabelEncoder()
train_df['Embarked_encoded'] = label_encoder_embark.fit_transform(train_df['Embarked'])
```

```python
train_df.drop(columns=["Sex", "Ticket", "Fare", "Cabin", "Embarked", "Age"], axis=1, inplace=True)

train_df.head()
```

```python
x = train_df.drop(columns=["Survived", "Name", "PassengerId"], axis=1)
y = train_df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
model = GaussianNB()

model.fit(x_train, y_train)

predictions = model.predict(x_test)

predictions
```

```python
precision = precision_score(y_test, predictions, average="macro")
recall = recall_score(y_test, predictions, average="macro")
f1 = f1_score(y_test, predictions, average="macro")
accuracy = accuracy_score(predictions, y_test)
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision (macro): {precision:.2f}")
print(f"Recall (macro): {recall:.2f}")
print(f"F1 Score (macro): {f1:.2f}")
```

```python
joblib.dump(model, 'linear_regression_model.joblib')
joblib.dump(label_encoder_embark, 'label_encoder_embark.joblib')
joblib.dump(label_encoder_sex, "label_encoder_sex.joblib")
```

```python
test_df = pd.read_csv("test.csv")

test_df.head(10)
```

```python
loaded_model = joblib.load('logistic_regression_model.joblib')
loaded_label_encoder_embark = joblib.load('label_encoder_embark.joblib')
loaded_label_encoder_sex = joblib.load("label_encoder_sex")
```

```python
test_df["Sex_encoded"] = loaded_label_encoder_sex.transform(test_df["Sex"])
test_df["Embarked_encoded"] = loaded_label_encoder_embark.transform(test_df["Embarked"])
```

```python
test_df.drop(columns=["Age", "Cabin", "Ticket", "Fare", "Sex", "Embarked"], axis=1, inplace=True)

test_df.head()
```

```python
feature_columns = ["PassengerId", "Pclass", "SibSp", "Parch", "Sex_encoded", "Embarked_encoded"]
```

```python
threshold = 0.5

survival_predictions = loaded_model.predict(test_df[feature_columns])

survival_predictions = (survival_predictions >= threshold).astype(int)
survival_predictions
```

```python
test_df["Survival_Prediction"] = survival_predictions

test_df.head()
```
