```python
import numpy as np
import pandas as  pd
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import accuracy_score, r2_score, precision_score, recall_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
df = pd.read_csv("train.csv")

df.head()
```

```python
df.groupby("Name")["SibSp"].sum().sort_values(ascending=False)
```

```python
df["Family_Size"] = df["SibSp"] +1
```

```python
df.head(20)
```

```python
df["Family_Name"] = df["Name"].str.split(',').str[0]
```

```python
df.head(20)
```

```python
df["Family_Name"].value_counts().sort_values(ascending=False)
```

```python
df.head(50)
```

```python
df["Cabin"].value_counts().sort_values(ascending=False)
```

```python
df.columns
```

```python
df.drop(["Fare", "Family_Name", "Ticket", "Cabin", "Age"], axis=1, inplace=True)
```

```python
df.head()
```

```python
label_encoder_sex = LabelEncoder()
df['Sex_Enc'] = label_encoder_sex.fit_transform(df['Sex'])

label_encoder_embark = LabelEncoder()
df['Embarked_Enc'] = label_encoder_embark.fit_transform(df['Embarked'])
```

```python
df.drop(["Sex", "Embarked"], axis=1, inplace=True)
```

```python
df.head()
```

```python
x = df.drop(["PassengerId", "Name", "Survived"], axis=1)
y = df["Survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.4)
```

```python
rfc = RandomForestClassifier()
```

```python
params = [{
    "n_estimators": [500, 1000, 1500],
    "criterion": ["entropy", "gini"],
    "min_samples_split": [5, 10, 15],
    "min_samples_leaf": [1, 2, 4],
    "max_depth": [10, 20, 30]
}]
```

```python
random_search = RandomizedSearchCV(rfc,
                                   params,
                                   cv=10,
                                   n_jobs= -1,
                                   scoring= "accuracy"
)
```

```python
random_search.fit(x_train, y_train)
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

```python
joblib.dump(random_search, "random_search.joblib")
joblib.dump(label_encoder_sex, "label_encoder_sex.joblib")
joblib.dump(label_encoder_embark, "label_encoder_embark.joblib")
```

```python
test = pd.read_csv("test.csv")

test.head()
```

```python
test["Family_Size"] = test["SibSp"] +1
```

```python
loaded_sex_encoder = joblib.load("label_encoder_sex.joblib")
loaded_embarked_encoder = joblib.load("label_encoder_embark.joblib")
loaded_random_search = joblib.load("random_search.joblib")
```

```python
test["Sex_Enc"] = loaded_sex_encoder.transform(test["Sex"])
test["Embarked_Enc"] = loaded_embarked_encoder.transform(test["Embarked"])
```

```python
test.drop(["Embarked", "Sex", "Ticket", "Fare", "Cabin", "Age"], axis=1, inplace=True)

test.head()
```

```python
feature_columns = ["Pclass", "SibSp", "Parch","Family_Size", "Sex_Enc", "Embarked_Enc"]
```

```python
threshold = 0.5

survival_predictions = loaded_random_search.predict(test[feature_columns])

survival_predictions = (survival_predictions >= threshold).astype(int)
survival_predictions
```

```python
test["Predictions"] = survival_predictions
```

```python
test.to_csv("Random_Search.csv")
```
